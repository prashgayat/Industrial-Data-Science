```python
# Importing important libraries
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Scatter Plot

# Create the data
x = [5, 7, 8,7, 2, 17, 2,
     9, 4, 11, 12, 9 ,6]

y = [99, 86, 87, 88, 100, 86,
     103, 87, 94, 78, 77, 85, 86]

# Setting the figure size (in inches)
plt.figure(figsize = (5,3))

#Plot
plt.scatter(x, y, color = 'red')

# Set the x and y labels
plt.xlabel('X  Values')
plt.ylabel('Y Values')
plt.title('Scatter Plot')
plt.show()

```

```python
# Line Plot
#Creating the data
year = [1920, 1930, 1940, 1950,1960, 1970, 1980, 1990, 2000, 2010, 2020]
unemployment_rate = [9.8, 12, 8, 7.2, 6.9, 7, 6.5, 6.2, 5.5, 6.3, 9.9]

#Plot
plt.figure(figsize = (5,3))
plt.plot(year, unemployment_rate, color = 'green', marker = 'o')
plt.xlabel('Year')
plt.ylabel('Unemployment Rate')
plt.title('Unemployment Rate Across years')
plt.show()
```

```python
#Bar Chart
#Creating the data
data = {
    'C' : 20,
    'C++' : 15,
    'Java' : 30,
    'Python' : 35
}

#Setting the figure size
plt.figure(figsize = (5,3))

# Plot
plt.bar(data.keys(), data.values(), color = 'black', width = 0.5)
plt.xlabel('Programming Languages')
plt.ylabel('Popularity')
plt.title('Popularity of Programming Languages')
plt.show()
```

```python
#Loading an inbuilt dataset
df = sns.load_dataset('tips')
df.head()
```

```python
# Doubt ->
print(df.shape)
```

```python
# counterplot()
# Setting the figure size
plt.figure(figsize = (5,3))

# Plot
sns.countplot(x = 'day', data = df)
plt.show()
```

```python
# counterplot()
# Setting the figure size
plt.figure(figsize = (5,3))

# Plot
sns.countplot(x = 'smoker', data = df)
plt.show()
```

```python
# Boxplot
plt.figure(figsize = (5,3))
plt.boxplot(df['total_bill'])
plt.title('Boxplot of Total Bill')
plt.show()
```

```python
#Histogram
plt.figure(figsize = (5,3))
plt.hist(df['total_bill'], bins = 30)
plt.title('Histogram of Total Bill')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```

```python
# Stacked Bar Chart
import numpy as np
# Creating the data
x = ['A', 'B', 'C', 'D']
y1 = np.array([10, 20, 10, 30])
y2 = np.array([20,2, 15, 25])
y3 = np.array([12, 15, 19, 6])
y4 = np.array([10, 25, 15, 20])

# Plot bars in stacked manner
plt.bar(x, y1, color = 'green')
plt.bar(x, y2, bottom = y1, color = 'red')
plt.bar(x, y3, bottom = y1 + y2, color = 'blue')
plt.bar(x, y4, bottom = y1 + y2 + y3, color = 'yellow')

# Set x and y  labels
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Stacked Bar Chart')
plt.legend(['Round 1', 'Round 2', 'Round 3', 'Round 4'])
plt.show()
```

```python
# Pie chart
# Creating the data
labels = ['Python', 'C++', 'Ruby', 'Java']
sizes = [215, 130, 245, 210]

#Plot
plt.figure(figsize = (5,3))
plt.pie(sizes, labels = labels, autopct = '%.2f%%', startangle = 120)
plt.title('Pie Chart')
plt.show()
```
